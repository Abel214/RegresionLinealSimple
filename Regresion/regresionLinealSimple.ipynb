```python
# ==============================
# REGRESIÓN LINEAL SIMPLE
# ==============================
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# 1. Preparacion de Datos
# Separacion de datos
X = np.array([1, 3, 4, 2, 1, 7]) #varible independiente
y = np.array([2, 3, 2.5, 2, 2, 3.5]) #variable objetivo
```

```python
#Promedios
x_mean = np.mean(X)
y_mean = np.mean(y)
```

```python
# Cálculo de b1 (pendiente) y b0 (intercepto)
b1 = np.sum((X - x_mean) * (y- y_mean)) / np.sum((X - x_mean)**2)
b0 = y_mean - b1 * x_mean
```

```python
print("Cálculo manual:")
print(f"b1 (pendiente) = {b1:.2f}")
print(f"b0 (intercepto) = {b0:.2f}")
print(f"Ecuación de regresión: Ŷ = {b0:.2f} + {b1:.2f}X")
```

```python
# --- Verificación con scikit-learn ---
X_reshaped = X.reshape(-1, 1)
# 2. Configuracion del modelo
# 3. Entrenamiento del modelo-- > .fit 
model=LinearRegression().fit(X_reshaped,y)
# 4. Prediccion del modelo
Y_pred=model.predict(X_reshaped)
```

```python
print("\nVerificación con scikit-learn:")
print(f"Pendidente (b1):{model.coef_[0]:.2f}")
print(f"Intercepto (b1):{model.intercept_:.2f}")
print(f"Ecuacione: Ŷ = {model.intercept_:.2f} + {model.coef_[0]:.2f}X")
# Evaluacion
print(f"R² = {r2_score(y, Y_pred):.2f}")
```

```python
#Gráfica
plt.scatter(X, y, color='blue', label='Datos regsitrados')
plt.plot(X,Y_pred, color='red', label='Recta de regresión')
plt.title('Regresión lineal simple: Nómina vs Ventas')
plt.xlabel('Nómina (X)')
plt.ylabel('Ventas (Y)')
plt.legend()
plt.grid(True)
plt.show()
```
